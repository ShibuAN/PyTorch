```python
%%time
value=0
for i in range(len(tensor)):
  value+=tensor[i]*tensor[i]
value
```

```python
%%time
torch.matmul(tensor,tensor)
```

```python
# Can also use the "@" symbol for matrix multiplication, though not recommended
%%time
tensor@tensor
```

There are two main rules that performing matrix multiplication needs to satisfy:
1. The **inner dimensions** must match
* `(3,2) @ (3,2)` won't work
* `(2,3) @ (3,2)` will work
* `(3,2) @ (2,3)` will work
2. The resulting matrix has the shape of the **outer dimensions**
* `(2,3) @ (3,2)` -> `(2,2)`
* `(3,2) @ (2,3)` -> `(3,3)`

###One of the most common error in deep learning: shape errors

```python
#shape for matrik multiplication
tensor_a=torch.tensor([[1,2],
                       [2,3],
                       [3,4]],dtype=torch.float32)
tensor_b=torch.tensor([[4,5],
                       [5,6],
                       [6,7]],dtype=torch.float32)
#torch.mm(tensor_a,tensor_b)
torch.matmul(tensor_a,tensor_b) # (this will error)
```

```python
tensor_a.shape,tensor_b.shape
```

To fix our tensor shape issues, we can manipulate the shape of one of our tensor using a **transpose**

A **transpose** switches the axes or dimensions of a gn tensor

```python
tensor_b.T, tensor_b
```

```python
torch.mm(tensor_a,tensor_b.T)
```

```python
torch.matmul(tensor_a,tensor_b.T)
```

##Reshaping,stacking,squeezing,unsqueezing tensors

* Reshaping-reshapes an i/p tensor to a defined shape.

* View-Return a view of an i/p tensor of certain shape but keep the same memory as the original tensor

* Stacking- combine multiple tensors on top of each other(vstack) or side by side (hstack)

* Squeeze- removes all `1` dim from a tensor

* Unsqueeze- add a `1`dim to a target tensor

* Permute-Return a view of the i/p with dims permuted(swapped) in a certain way


```python
#let's create a tensor
x=torch.arange(1.0,9.0)
x, x.shape
```

```python
#add a extra dim
x_reshaped=x.reshape(2,4)
x_reshaped
```

```python
#change the view
z=x.view(2, 4)
z, z.shape
```

```python
#changing z changes x (because a view of a tensor shares the same memory as the original shape)
z[:,0]=5
z,x
```

```python
#stack tensors on top of each other-hstack
x_stacked=torch.stack([x,x,x,x],dim=0)
x_stacked
```

```python
#vstack
x_stacked=torch.stack([x,x,x,x],dim=1)
x_stacked
```

```python
#torch.squeeze()-removes all single dims from a target tensor
x_reshaped
```

```python
x_reshaped.shape
```

```python
x_reshaped.squeeze()
```

```python
x_reshaped.squeeze().shape
```

```python
x_reshaped=x_reshaped.reshape(8,1)
x_reshaped
```

```python
x_reshaped.shape
```

```python
x_reshaped.squeeze()
```

```python
x_reshaped.squeeze().shape
```

```python
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
#torch.unsqueeze()-add a single dims to a target tensor at a specific dims
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

#Add a extra dim with unsqueeze
x_unsqueezed=x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"\nNew shape: {x_unsqueezed.shape}")
```

```python
#torch-permute-rearranges the dims of a target tensor in a specified order
x_original=torch.rand(size=(224,224,3)) #[height,width,colour_channels]

#Permute the original tensor to rearrange the axis (or dim) order
x_permuted=x_original.permute(2,0,1) #shifts axis 0->1, 1->2, 2->0
print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

##Reproducbility(trying to take random out of random)
In short how a neural network learns:

`start with random numbers-> tensor operations -> update random numbers to try and make them of the data -> again -> again -> again...`

To reduce the randomness in neural networks and pyTorch comes the concept of a **random seed.**

Essentially what the random seed does is "flavour" the randomness.

```python
torch.rand(3,3)
```

```python
#create a new random tensors
random_tensor_a=torch.rand(3,5)
random_tensor_b=torch.rand(3,5)

print(random_tensor_a)
print(random_tensor_b)
print(random_tensor_a==random_tensor_b)
```

```python
#Let's make some random but reproduciable tensor
#set the random seed
Random_seed=42
torch.manual_seed(Random_seed)
random_tensor_c=torch.rand(3,5)

torch.manual_seed(Random_seed)
random_tensor_d=torch.rand(3,5)

print(random_tensor_c)
print(random_tensor_d)
print(random_tensor_c==random_tensor_d)
```
